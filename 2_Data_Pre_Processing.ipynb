```python
# Pivoting the data.. from KPI column to individual columns with quarter
df = df_qtrly_sales
# Pivotted the pivot
pivot_columns = list(set(df.columns) - set(['KPI','Tikr']))  
df_pivot = pd.pivot_table(df, index=['Tikr'], columns=['KPI'], values=pivot_columns, aggfunc='sum')
# Replacing spaces from the column names and joining the two layer of column names into one.
df_pivot.columns =[str(s2) + str('_') + s1  for (s1,s2) in df_pivot.columns.tolist()]
df_pivot.reset_index(inplace=True)
df_pivot.columns = df_pivot.columns.str.replace('+', '')
df_pivot.columns = df_pivot.columns.str.replace(' ', '_')
# Reordering columns by name so that all similar column get together
df_pivot = df_pivot.reindex(sorted(df_pivot.columns), axis=1)
df_pivot.head(4)
```

```python
# disable chained assignments
pd.options.mode.chained_assignment = None 

def percentage_change(col1,col2):
    try:
        out = ((col2 - col1) / col1) * 100
    except:
        out = np.nan
    return out
    #return ((df_sales.loc[:,col2] - df_sales.loc[:,col1])/df_sales.loc[:,col1]) * 100

quarter_cols = [['2019_Q1_QoQ','2018_Q4','2019_Q1'],
             ['2019_Q2_QoQ','2019_Q1','2019_Q2'],
             ['2019_Q3_QoQ','2019_Q2','2019_Q3'],
             ['2019_Q4_QoQ','2019_Q3','2019_Q4'],
             ['2020_Q1_QoQ','2019_Q4','2020_Q1'],
             ['2020_Q2_QoQ','2020_Q1','2020_Q2'],
             ['2020_Q3_QoQ','2020_Q2','2020_Q3'],
             ['2020_Q4_QoQ','2020_Q3','2020_Q4'],
             ['2021_Q1_QoQ','2020_Q4','2021_Q1'],
             ['2021_Q2_QoQ','2021_Q1','2021_Q2'],
             ['2021_Q3_QoQ','2021_Q2','2021_Q3']]

def calc_percentage_diff(df_var,var,col_out,col1,col2):
    col1 = str(var) + str('_') + str(col1)
    col2 = str(var) + str('_') + str(col2)
    col_out = str(var) + str('_') + str(col_out)
    #print(col_out,col1,col2)
    df_var[col_out] = percentage_change(df_var[col1],df_var[col2])
    #print(df_sales[col_out])

var = 'Sales'
#vars = ['Sales','Expenses','','']
for combination in quarter_cols:
    var_out = combination[0]
    var1 = combination[1]
    var2 = combination[2]
    calc_percentage_diff(df_sales,var,var_out,var1,var2)
```

```python
#Throught this loop we will process multiple colmns at once
# We will clean the variables and also calculate the QoQ for each variable

variables = ['Sales','Revenue','Expenses','Net_Profit','Operating_Profit','EPS_in_Rs','Tax_','OPM_']
   
df_quarter = pd.DataFrame()
for var in variables:
    print(var)
    #matching = [s for s in df_pivot.columns if "Sales\xa0" in s]        # Getting all the columns containing 'sales/profit etc'
    matching = [s for s in df_pivot.columns if var in s]        # Getting all the columns containing 'sales/profit etc'
    matching.append('Tikr')                                             # Added tikr column
    matching = matching[-1:] + matching[:-1]                            # Moving Tikr column from bottom to top of the list
    #print(matching)
    df_var = df_pivot[matching]
    #print(df_sales.head(4))
    df_var_col = df_var.columns
    df_var_col = [i.replace('\xa0','') for i in df_var_col]         # Getting rid of special characters 
    print(df_var_col)
    print('Number of cols in df - ',len(df_var_col))
    df_var.columns = df_var_col

    # Calculating QoQ calculations over all available quarter columns
    # List of output and input columns is in quarter_cols
    # Updating each QoQ column 1 by 1
    for combination in quarter_cols:
        try:    
            var_out = combination[0]
            var1 = combination[1]
            var2 = combination[2]
            calc_percentage_diff(df_var,var,var_out,var1,var2)
        except Exception as e:
            print('Error - ',e)
    df_quarter = pd.concat([df_quarter.reset_index(drop=True), df_var], axis=1)
```

```python
# disable chained assignments
pd.options.mode.chained_assignment = None 

def percentage_change(col1,col2):
    try:
        out = ((col2 - col1) / col1) * 100
    except:
        out = np.nan
    return out
    #return ((df_sales.loc[:,col2] - df_sales.loc[:,col1])/df_sales.loc[:,col1]) * 100

annual_cols =  [['2011_YoY','2010','2011'],
                ['2012_YoY','2011','2012'],
                ['2013_YoY','2012','2013'],
                ['2014_YoY','2013','2014'],
                ['2015_YoY','2014','2015'],
                ['2016_YoY','2015','2016'],
                ['2017_YoY','2016','2017'],
                ['2018_YoY','2017','2018'],
                ['2019_YoY','2018','2019'],
                ['2020_YoY','2019','2020'],
                ['2021_YoY','2021','2021']]


def calc_percentage_diff(df_var,var,col_out,col1,col2):
    col1 = str(var) + str('_') + str(col1)
    col2 = str(var) + str('_') + str(col2)
    col_out = str(var) + str('_') + str(col_out)
    #print(col_out,col1,col2)
    df_var[col_out] = percentage_change(df_var[col1],df_var[col2])
    return df_var
    #print(df_sales[col_out])
```

```python
#Throught this loop we will process multiple colmns at once
# We will clean the variables and also calculate the YoY for each variable
def flatten_dataframe(df_pivot,variables):   
    df_out = pd.DataFrame()
    for var in variables:
        print(var)
        #matching = [s for s in df_pivot.columns if "Sales\xa0" in s]        # Getting all the columns containing 'sales/profit etc'
        matching = [s for s in df_pivot.columns if var in s]        # Getting all the columns containing 'sales/profit etc'
        matching.append('Tikr')                                             # Added tikr column
        matching = matching[-1:] + matching[:-1]                            # Moving Tikr column from bottom to top of the list
        #print(matching)
        df_var = df_pivot[matching]
        #print(df_sales.head(4))
        df_var_col = df_var.columns
        df_var_col = [i.replace('\xa0','') for i in df_var_col]         # Getting rid of special characters 
        print(df_var_col)
        print('Number of cols in df - ',len(df_var_col))
        df_var.columns = df_var_col

        # Calculating QoQ calculations over all available quarter columns
        # List of output and input columns is in quarter_cols
        # Updating each QoQ column 1 by 1
        for combination in annual_cols:
            try:    
                var_out = combination[0]
                var1 = combination[1]
                var2 = combination[2]
                df_var = calc_percentage_diff(df_var,var,var_out,var1,var2)
            except Exception as e:
                print('Error - ',e)
        df_out = pd.concat([df_out.reset_index(drop=True), df_var], axis=1)
    return df_out

bal_variables = ['Other_Assets', 'Total_Liabilities', 'Share_Capital', 'Investments', 'Borrowings', 'Reserves', 'Total_Assets', 'Fixed_Assets', 'Other_Liabilities', 'CWIP']
cash_variables = ['Cash_from_Investing_Activity', 'Cash_from_Operating_Activity', 'Cash_from_Financing_Activity', 'Net_Cash_Flow']
ratio_variables = ['Days_Payable', 'ROCE_', 'Cash_Conversion_Cycle', 'Working_Capital_Days', 'ROE_', 'Inventory_Days', 'Debtor_Days']

df_balance_sheet_flat = flatten_dataframe(df_balance_sheet,bal_variables)
df_cash_flow_flat = flatten_dataframe(df_cash_flow,cash_variables)
df_ratio_flat = flatten_dataframe(df_ratio,ratio_variables)
```
