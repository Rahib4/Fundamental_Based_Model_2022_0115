```python
df_bhavcopy_yfin = pd.read_csv('df_bhavcopy_yfin.csv')
df_bhavcopy_melt = pd.melt(df_bhavcopy_yfin, id_vars=["Date"])
df_bhavcopy_melt = df_bhavcopy_melt[df_bhavcopy_melt['value'].notna()]
df_bhavcopy_melt[['KPI','Tikr']] = df_bhavcopy_melt['variable'].str.split('_',1, expand=True)
df_bhavcopy_melt = df_bhavcopy_melt.drop(['variable'],axis=1)
df_bhavcopy_melt.columns = ['TIMESTAMP','CLOSE','KPI','SYMBOL']

df_bhavcopy_melt
```

```python
# #df_nifty.shape
# df_nifty = data.reset_index()
# df_nifty.columns = ['TIMESTAMP','OPEN','HIGH','LOW','CLOSE','ADJ_CLOSE','TOTTRDQTY']
# df_nifty.head()
df_bhavcopy_melt['TIMESTAMP'] = pd.DatetimeIndex(df_bhavcopy_melt['TIMESTAMP']).strftime('%d-%b-%Y').str.upper()
df_bhavcopy_melt.head()
```

```python
df_bhav_pivot = pd.pivot_table(df_bhavcopy_melt,index=['TIMESTAMP','SYMBOL'], columns=['KPI'], values=['CLOSE'], aggfunc='sum')
df_bhav_pivot = df_bhav_pivot.reset_index()
df_bhav_pivot.columns = ['_'.join(col) for col in df_bhav_pivot.columns]
df_bhav_pivot.columns = ['TIMESTAMP','SYMBOL','CLOSE','Volume']
df_bhav_pivot
```
