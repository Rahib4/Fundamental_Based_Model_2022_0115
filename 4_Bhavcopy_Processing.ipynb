```python
#Getting average of the 4 values
# Earlier we used to have split adjustments separately, since we shifted to yfinance so we dont that anymore
df_bhav_split_adjusted = df_bhavcopy.copy()
#df_bhav_split_adjusted['AVG'] = df_bhav_split_adjusted[['CLOSE']].mean(axis=1)

df_bhav_pivot = pd.pivot_table(df_bhav_split_adjusted, index=['SYMBOL'], columns=['Yr_Month'], values=['CLOSE'], aggfunc='mean')
# Replacing spaces from the column names and joining the two layer of column names into one.
df_bhav_pivot.columns =[str(s2) + str('_') + s1  for (s1,s2) in df_bhav_pivot.columns.tolist()]
df_bhav_pivot.reset_index(inplace=True)
# Reordering columns by name so that all similar column get together
df_bhav_pivot = df_bhav_pivot.reindex(sorted(df_bhav_pivot.columns), axis=1)
df_bhav_pivot.head(3)
```

```python
df_bhav_yearly = pd.DataFrame()
for i in range(2009,2022):
    #print(i)
    df_bhav_yearly[i] = df_bhav_pivot.filter(regex=str(i)).mean(axis=1)
df_bhav_yearly['Tikr'] = df_bhav_pivot['SYMBOL']
df_bhav_yearly.head(10).to_clipboard()
df_bhav_yearly.head()
```

```python
len(df_bhav_yearly.columns) - 2
cols = df_bhav_yearly.columns
# Percentae calclation
def percentage_change(col1,col2):
    try:
        out = ((col2 - col1) / col1) * 100
    except:
        out = np.nan
    return out
    #return ((df_sales.loc[:,col2] - df_sales.loc[:,col1])/df_sales.loc[:,col1]) * 100

df_bhav_delta = pd.DataFrame()

# Looping over columns to get the percentage change YoY
# Earlier we were using 3 yrs delta for model
# Shifting to 1 yr return now
for i in range(len(df_bhav_yearly.columns) - 2):
    #print(cols[i])
    if i >= 2:
        df_bhav_delta[cols[i]] = percentage_change(df_bhav_yearly[cols[i]],df_bhav_yearly[cols[i+2]])

df_bhav_delta['Tikr'] = df_bhav_yearly['Tikr'].str.replace('^NSEI','Nifty50')
df_bhav_delta.iloc[-1,len(df_bhav_delta.columns)-1] = 'Nifty50'
#df_bhav_delta['Tikr'] = df_bhav_delta['Tikr'].str.replace('.NS','')
df_bhav_delta['Tikr'] = df_bhav_delta['Tikr'].str[:-3]
df_bhav_delta.head(4)
#df_bhav_delta.tail(4)
```

```python
# Getting MoM return trend for each ticker
df_bhav_pivot.columns
df_bhav_monthly = pd.DataFrame()
# for i in range(2014,2022):
#     #print(i)
#     df_bhav_monthly[i] = df_bhav_pivot.filter(regex=str(i)).mean(axis=1)
# df_bhav_yearly['Tikr'] = df_bhav_pivot['SYMBOL']
# df_bhav_yearly.head(10).to_clipboard()
# df_bhav_yearly.head()

def percentage_change(col1,col2):
    try:
        out = ((col2 - col1) / col1) * 100
    except:
        out = np.nan
    return out
    
cols = df_bhav_pivot.columns
def monthly_return(df):
    for i in range(len(df_bhav_pivot.columns) - 1):
        #print(cols[i])
        if i >= 1:
            df_bhav_monthly[cols[i]] = percentage_change(df[cols[i-1]],df[cols[i]])
    return df_bhav_monthly

df_bhav_monthly_delta = monthly_return(df_bhav_pivot)
df_bhav_monthly_delta['Tikr'] = df_bhav_pivot['SYMBOL'].str.replace('^NSEI','Nifty50')
#df_bhav_monthly_delta['Tikr'] = df_bhav_monthly_delta['Tikr'].str.replace('^NSEI','Nifty50')
df_bhav_monthly_delta.iloc[-1,len(df_bhav_monthly_delta.columns)-1] = 'Nifty50'
df_bhav_monthly_delta['Tikr'] = df_bhav_monthly_delta['Tikr'].str[:-3]
df_bhav_monthly_delta.to_csv('df_bhav_monthly_delta.csv')
```
